```python
import pandas as pd

car_data = pd.read_csv("../vehicles_us.csv")
car_data.head()
```

### 1. Histograma 
Visualización en browser

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = 'browser'

fig = px.histogram(car_data, x='odometer', nbins=50, title='Distribución del odómetro')
fig.show()
```

### 3. Gráfica de Dispersión
Visualización en browser

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = 'browser'

fig = px.scatter(
    car_data, 
    x='odometer', 
    y='price',
    title='Precio vs. Odómetro'
)
fig.show()
```

```python
import streamlit as st
import pandas as pd
import plotly.express as px
import plotly.io as pio

# Always render charts in browser
pio.renderers.default = "browser"

# Load data
car_data = pd.read_csv("vehicles_us.csv")

# Title
st.title("Car Sales S7 - Análisis Interactivo")

st.write("Selecciona una opción para visualizar los datos del dataset.")

# Buttons
if st.button("Mostrar histograma del odómetro"):
    fig = px.histogram(car_data, x="odometer", nbins=50, title="Distribución del odómetro")
    st.write("Histograma generado:")
    st.plotly_chart(fig)

if st.button("Mostrar diagrama de dispersión: precio vs odómetro"):
    fig2 = px.scatter(car_data, x="odometer", y="price", title="Precio vs Odómetro")
    st.write("Gráfico de dispersión generado:")
    st.plotly_chart(fig2)
```
